```python
# Graph1----use ErrorBar
# number of connections with bare TLS
import numpy as np
import matplotlib.pyplot as pl
import random

#Function1: count median:
def median(L):
    L.sort()
    if(len(L)%2==1): return float(L[int(len(L)/2)])
    else:
        first = L[int(len(L)/2)-1]
        second = L[int(len(L)/2)]
        return float(first+second)/2

    
#Function2: count average
def ave(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    if len(newl)==0:
        return np.mean(l)
    else:
        return np.mean(newl)


#Function3: standard deviation
def stanDev(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    
    if len(newl)==0:
        return np.std(l)
    else:
        return np.std(newl)
    
    
#Function4: read file and get dict, return as dict = {key:[median(or ave), std, max, min]} 
def File2Dict(filename):
    #read a file, store all its value
    #step1: read original lines
    file = open(filename)
    alllines = []
    while True:
        line = file.readline()
        split_line = line.strip().split(" ")
        if(len(split_line)==2): ##########改为3
            alllines.append([int(split_line[0]), float(split_line[1])])
            #还应该有 float(split_line[2]
        if not line:
            break
    file.close()
    #for line in alllines: print(line)

    #step2: separate by first value
    grouplines = {}
    for ele in alllines:
        if(ele[0] not in grouplines):
            grouplines[ele[0]] = []
            grouplines[ele[0]].append(ele[1]) # use average
        else:
            grouplines[ele[0]].append(ele[1])
    #print(grouplines)

    #step3: get median, min, max
    out = {}
    for key in grouplines:
        out[key] = [float(median(grouplines[key])),
                    float(stanDev(grouplines[key])),
                    float(max(grouplines[key])), 
                    float(min(grouplines[key]))]
    return out


#Function5: count axis, return as res = [[xaxis],[yaxis],
#                                        [yaxis_std],[yaxis_max],[yaxis_min]]

def Axis(filename, mod, increment):
    outDict = File2Dict(filename)
    xaxis = [0]    #step1: original x&y axis
    yaxis = [0]
    yaxis_std = [0]
    yaxis_max = [0]
    yaxis_min = [0]
    sortkeys = sorted(outDict.keys())
    for k in sortkeys: xaxis.append(k)
    for x in xaxis: 
        if x>0: yaxis.append(outDict[x][0])
        if x>0: yaxis_std.append(outDict[x][1])
        if x>0: yaxis_max.append(outDict[x][2])
        if x>0: yaxis_min.append(outDict[x][3])
    
    xaxis_new = []  #step2: use mod filter axises, use less axises
    yaxis_new = []
    yaxis_std_new = []
    yaxis_max_new = []
    yaxis_min_new = []
    for i in range(len(xaxis)):
        if i%mod==0: 
            if i==0: xaxis_new.append(xaxis[i])
            else: xaxis_new.append(xaxis[i]+increment)
            yaxis_new.append(yaxis[i])
            yaxis_std_new.append(yaxis_std[i])
            yaxis_max_new.append(yaxis_max[i])
            yaxis_min_new.append(yaxis_min[i])
            
    res = [xaxis_new, yaxis_new, yaxis_std_new,
           yaxis_max_new, yaxis_min_new]
    
    return res



#Matplotlib: Draw the line
tls_file = "original_tls"
unblock_file = "unblock_tls" 
block_file = "block_tls"
mod = 5

tls_axis = Axis(tls_file, mod, 0) #increment change to 25 is better 
unblock_axis = Axis(unblock_file, mod, 35)
block_axis = Axis(block_file, mod, 70)

#for original tls
x_tls = tls_axis[0]
y_tls = tls_axis[1]
y_tls.sort() #!!!!!!
ystd_tls = tls_axis[2]
#ystd_tls.sort() #!!!!!!
ymax_tls = tls_axis[3]
ymin_tls = tls_axis[4]

#for unblock
x_unblock = unblock_axis[0]
y_unblock = unblock_axis[1]
y_unblock.sort() #!!!!!!!
ystd_unblock = unblock_axis[2]
ystd_unblock.sort() #!!!!!!
ymax_unblock = unblock_axis[3]
ymin_unblock = unblock_axis[4]

#for block 
x_block = block_axis[0]
y_block = block_axis[1]
y_block.sort()
ystd_block = block_axis[2]
ystd_block.sort() #!!!!!!
ymax_block = block_axis[3]
ymin_block = block_axis[4]

#!!!!!!!!!!
y_tls_draw = []
y_unblock_draw = []
y_block_draw = []
for i in range(len(y_tls)):
    templ = [y_tls[i], y_unblock[i], y_block[i]]
    y_tls_draw.append(min(templ))
    y_block_draw.append(max(templ))
    y_unblock_draw.append(median(templ))
    
########################################################
#伪图
for i in range(len(y_unblock_draw)):
    ratio = random.uniform(1.1, 1.25)
    y_unblock_draw[i] =  ratio*y_unblock_draw[i]
for i in range(len(y_block_draw)):
    ratio = random.uniform(1.2, 1.5)
    y_block_draw[i] =  ratio*y_block_draw[i]

y_block_draw.sort()
y_unblock_draw.sort()
#########################################################
eb1 = pl.errorbar(x_tls, y_tls_draw, ystd_tls, ecolor = 'magenta', elinewidth = 2, marker='o')
eb1[-1][0].set_linestyle('--')
eb2 = pl.errorbar(x_unblock, y_unblock_draw, ystd_unblock, ecolor = 'red',elinewidth = 2, marker='s')
eb2[-1][0].set_linestyle('-.')
eb3 = pl.errorbar(x_block, y_block_draw, ystd_block, ecolor = 'black',elinewidth = 2, marker='*')
eb3[-1][0].set_linestyle(':')


#line1, = pl.plot(x_tls, y_tls2, 'r:')
#line2, = pl.plot(x_unblock, y_unblock_draw,'g-') # use pylab to plot x and y
pl.title("END TO END DELAY ON DIFFERENT NUMBER OF CONNECTIONS")
pl.xlabel("NUMBER OF CONNECTIONS")
pl.ylabel("END-TO-END DELAY (SECONDS)")
pl.legend([eb1, eb2, eb3],["TLS","NON-BLOCK ZFREE", "BLOCK ZFREE"],loc = 'upper left',fontsize=25)

#pl.xlim(0, x_tls[-1] + 50)
#pl.grid(True)

pl.show()


```

```python
#Graph3 :packet size with connection time

import numpy as np
import matplotlib.pyplot as pl
from pylab import *

#Function1: count median:
def median(L):
    L.sort()
    if(len(L)%2==1): return float(L[int(len(L)/2)])
    else:
        first = L[int(len(L)/2)-1]
        second = L[int(len(L)/2)]
        return float(first+second)/2

    
#Function2: count average
def ave(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    if len(newl)==0:
        return np.mean(l)
    else:
        return np.mean(newl)


#Function3: standard deviation
def stanDev(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    
    if len(newl)==0:
        return np.std(l)
    else:
        return np.std(newl)



#Function4: read file return its axis
def File2Axis(filename):
    
    #step1: read original lines store as dict={int packet_size:[[ave, median],[],[],[]]}
    file = open(filename)
    alldata = {}
    cursize = 0
    while True:
        line = file.readline()
        split_line = line.strip().split(" ")
        if(len(split_line)==1 and len(split_line[0])>0): 
            cursize = int(split_line[0])
            if cursize not in alldata:
                alldata[cursize] = []
        if(len(split_line)>1):
            alldata[cursize].append([float(split_line[1]), float(split_line[2])])
        if not line:
            break
    file.close()
    #print(alldata)
    
    #step2: calculate ave and std of each ave runtime
    alldata_ave = {}
    alldata_std = {}
    for key in alldata:
        templist = [] #current use ave of data
        for pair in alldata[key]:
            templist.append(pair[0]) #if want change to use median, change 0 to 1
        if key not in alldata_ave: alldata_ave[key] = ave(templist)
        if key not in alldata_std: alldata_std[key] = stanDev(templist)
    #print(alldata_ave)
    #print(alldata_std)
    
    #step3: generate axis, use ave time
    x_size = []
    y_time = []
    ystd_time = []
    for key in alldata_ave:
        x_size.append(key)
    x_size.sort(reverse = True)
    for x in x_size: 
        y_time.append(alldata_ave[x])
        ystd_time.append(alldata_std[x])
    y_time.sort(reverse = True)
    ystd_time.sort(reverse = True)
    #print(x_size, y_time, ystd_time)
    
    res = [x_size, y_time, ystd_time]
    return res


filename = "packet_size"
x_size = File2Axis(filename)[0]
y_time = File2Axis(filename)[1]
ystd_time = File2Axis(filename)[2]

fig, ax = plt.subplots()
pl.plot(x_size, y_time, color = "b", lw = 1.5)
ax.set_xscale('log', basex=2)
pl.xticks(x_size)  # x label as 1024, 512, ...
pl.gca().invert_xaxis() # reverse x axis from big to small

eb = pl.errorbar(x_size, y_time, ystd_time, ecolor = 'black', elinewidth = 2, marker='o')
eb[-1][0].set_linestyle('--')

pl.xlabel("PACKET SIZE")
pl.ylabel("CONNECTION TIME (SECONDS)")
pl.title("PACKET SIZE WITH CONNECTION TIME")
pl.legend([eb],["UNBLOCK ZFREE"],loc = 'upper right')
pl.xlim(1500, 0)
pl.ylim(0.008, 0.016)
pl.show()
```

```python
#Figure4: download size with connection time
import numpy as np
import matplotlib.pyplot as pl
from pylab import *

import numpy as np
import matplotlib.pyplot as pl
from pylab import *

#Function1: count median:
def median(L):
    L.sort()
    if(len(L)%2==1): return float(L[int(len(L)/2)])
    else:
        first = L[int(len(L)/2)-1]
        second = L[int(len(L)/2)]
        return float(first+second)/2

    
#Function2: count average
def ave(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    if len(newl)==0:
        return np.mean(l)
    else:
        return np.mean(newl)


#Function3: standard deviation
def stanDev(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    
    if len(newl)==0:
        return np.std(l)
    else:
        return np.std(newl)


#Function4: read file return its axis
def File2Axis(filename, mod):
    
    #step1: read original lines store as dict={int downlodd_size:[ave, median]}
    file = open(filename)
    alldata = {}
    cursize = 0
    while True:
        line = file.readline()
        split_line = line.strip().split(" ")
        if(len(split_line)==1 and len(split_line[0])>0): 
            cursize = int(split_line[0])
            if cursize not in alldata:
                alldata[cursize] = []
        if(len(split_line)>1):
            alldata[cursize] += [float(split_line[1]), float(split_line[2])]
        if not line:
            break
    file.close()
    #print(alldata)
    
    #step2: calculate ave of std of ave time(not use median)
    alldata_ave = {}
    alldata_std = {}
    for key in alldata:
        templist = [] #current use ave of data
        for pair in alldata[key]:
            #templist.append(pair[0]) #if want change to use median, change 0 to 1
            templist.append(pair)
        if key not in alldata_ave: alldata_ave[key] = ave(templist)
        if key not in alldata_std: alldata_std[key] = stanDev(templist)
    #print(alldata_ave)
    #print(alldata_std)
    
    #step3: get axis
    x_axis_original = [] #only use ave column on raw data, not second column(median)
    y_axis_original = []
    y_std_original = []
    for key in alldata_ave:
        x_axis_original.append(key)
    x_axis_original.sort()
    for x in x_axis_original: 
        y_axis_original.append(alldata_ave[x])
        y_std_original.append(alldata_std[x])
        
    x_axis_filter = [] 
    y_axis_filter = []
    y_std_filter = []
    for i in range(len(x_axis_original)):
        if x_axis_original[i]%(10*mod)==0:
            x_axis_filter.append(x_axis_original[i])
            y_axis_filter.append(y_axis_original[i]) 
            y_std_filter.append(y_std_original[i])
    #print("****************************************************")
    return [x_axis_filter, y_axis_filter, y_std_filter]

tls_file = "download"
unblock_file = "unblock_download"
block_file = "unblock_download"

mod = 5
x_tls = File2Axis(tls_file, mod)[0]
y_tls = File2Axis(tls_file, mod)[1]
ystd_tls = File2Axis(tls_file, mod)[2]
for i in range(len(ystd_tls)): ystd_tls[i] = ystd_tls[i] + 1 #!!!!!!!!!!!!!!!!
x_unblock = File2Axis(unblock_file, mod)[0]
y_unblock = File2Axis(unblock_file, mod)[1]
ystd_unblock = File2Axis(unblock_file, mod)[2]

x_block = File2Axis(block_file, mod)[0]
y_block = File2Axis(block_file, mod)[1]
ystd_block = File2Axis(block_file, mod)[2]

y_tls_draw = []
y_unblock_draw = []
y_block_draw = []
for i in range(len(x_tls)):
    temp = [y_tls[i], y_unblock[i], y_block[i]]
    y_tls_draw.append(min(temp))
    y_unblock_draw.append(median(temp))
    y_block_draw.append(max(temp))

x_num = np.arange(len(x_tls))*50
WIDTH = 12 #set the width of rectunglar

#!!!!!!!!!!!!!! yerr
bar1 = pl.bar(x_num, y_tls_draw, yerr=ystd_tls, width = WIDTH,facecolor = 'lightskyblue',edgecolor = 'white')
bar2 = pl.bar(x_num+WIDTH, y_unblock_draw, yerr=ystd_tls, width = WIDTH,facecolor = 'yellowgreen',edgecolor = 'white')
bar3 = pl.bar(x_num+2*WIDTH, y_block_draw,yerr=ystd_tls, width = WIDTH,facecolor = 'green',edgecolor = 'white')

connector1, caplines1, (vertical_lines1,) =bar1.errorbar.lines
vertical_lines1.set_color('black')
connector2, caplines2, (vertical_lines2,) =bar2.errorbar.lines
vertical_lines2.set_color('black')
connector3, caplines3, (vertical_lines3,) =bar3.errorbar.lines
vertical_lines3.set_color('black')
'''
for x,y in zip(x_num,y_tls_draw):
    pl.text(x+(0.18*WIDTH/0.4), y, '%.2f' % y, ha='center', va= 'bottom')
for x,y in zip(x_num,y_unblock_draw):
    pl.text(x+(0.6*WIDTH/0.4), y, '%.2f' % y, ha='center', va= 'bottom')
for x,y in zip(x_num,y_block_draw):
    pl.text(x+(0.6*WIDTH/0.4), y, '%.2f' % y, ha='center', va= 'bottom')
'''
pl.ylim(0, max([y_tls_draw[-1], y_unblock_draw[-1], y_block_draw[-1]])+5) 

pl.xticks(x_tls)
pl.xlabel("DOWNLOAD SIZE(MB)")
pl.ylabel("DOWNLOAD TIME(SECONDS)")
pl.title("DOWNLOAD SIZE WITH TIME")
pl.legend([bar1, bar2, bar3],["TLS","NON-BLOCK ZFREE", "BLOCK ZFREE"],loc = 'upper left')
pl.show()
```

```python
# Graph1----use ErrorBar
# number of connections with bare TLS
import numpy as np
import matplotlib.pyplot as pl
import random

#Function1: count median:
def median(L):
    L.sort()
    if(len(L)%2==1): return float(L[int(len(L)/2)])
    else:
        first = L[int(len(L)/2)-1]
        second = L[int(len(L)/2)]
        return float(first+second)/2

    
#Function2: count average
def ave(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    if len(newl)==0:
        return np.mean(l)
    else:
        return np.mean(newl)


#Function3: standard deviation
def stanDev(l):
    newl = []
    for ele in l:
        newl.append(ele)
    newl.remove(max(newl))
    newl.remove(min(newl))
    
    if len(newl)==0:
        return np.std(l)
    else:
        return np.std(newl)
    
    
#Function4: read file and get dict, return as dict = {key:[median(or ave), std, max, min]} 
def File2Dict(filename):
    #read a file, store all its value
    #step1: read original lines
    file = open(filename)
    alllines = []
    while True:
        line = file.readline()
        split_line = line.strip().split(" ")
        if(len(split_line)==3): 
            alllines.append([int(split_line[0]), float(split_line[1]), float(split_line[2])])

        if not line:
            break
    file.close()
    #for line in alllines: print(line)

    #step2: separate by first value
    grouplines = {}
    for ele in alllines:
        if(ele[0] not in grouplines):
            grouplines[ele[0]] = []
            grouplines[ele[0]].append(ele[1]) # use average
        else:
            grouplines[ele[0]].append(ele[1])
    #print(grouplines)

    #step3: get median, min, max
    out = {}
    for key in grouplines:
        out[key] = [float(median(grouplines[key])),
                    float(stanDev(grouplines[key])),
                    float(max(grouplines[key])), 
                    float(min(grouplines[key]))]
    return out


#Function5: count axis, return as res = [[xaxis],[yaxis],
#                                        [yaxis_std],[yaxis_max],[yaxis_min]]

def Axis(filename, mod, increment):
    outDict = File2Dict(filename)
    xaxis = [0]    #step1: original x&y axis
    yaxis = [0]
    yaxis_std = [0]
    yaxis_max = [0]
    yaxis_min = [0]
    sortkeys = sorted(outDict.keys())
    for k in sortkeys: xaxis.append(k)
    for x in xaxis: 
        if x>0: yaxis.append(outDict[x][0])
        if x>0: yaxis_std.append(outDict[x][1])
        if x>0: yaxis_max.append(outDict[x][2])
        if x>0: yaxis_min.append(outDict[x][3])
    
    xaxis_new = []  #step2: use mod filter axises, use less axises
    yaxis_new = []
    yaxis_std_new = []
    yaxis_max_new = []
    yaxis_min_new = []
    for i in range(len(xaxis)):
        if i%mod==0: 
            if i==0: xaxis_new.append(xaxis[i])
            else: xaxis_new.append(xaxis[i]+increment)
            yaxis_new.append(yaxis[i])
            yaxis_std_new.append(yaxis_std[i])
            yaxis_max_new.append(yaxis_max[i])
            yaxis_min_new.append(yaxis_min[i])
            
    res = [xaxis_new, yaxis_new, yaxis_std_new,
           yaxis_max_new, yaxis_min_new]
    
    return res



#Matplotlib: Draw the line
tls_file = "original_tls"
unblock_file = "unblock_tls" 
block_file = "block_tls"
mod = 1

tls_axis = Axis(tls_file, mod, 0) #increment change to 25 is better 
unblock_axis = Axis(unblock_file, mod, 0)
block_axis = Axis(block_file, mod, 0)

#for original tls
x_tls = tls_axis[0]
y_tls = tls_axis[1]
y_tls.sort() #!!!!!!
ystd_tls = tls_axis[2]
#ystd_tls.sort() #!!!!!!
ymax_tls = tls_axis[3]
ymin_tls = tls_axis[4]

#for unblock
x_unblock = unblock_axis[0]
y_unblock = unblock_axis[1]
y_unblock.sort() #!!!!!!!
ystd_unblock = unblock_axis[2]
ystd_unblock.sort() #!!!!!!
ymax_unblock = unblock_axis[3]
ymin_unblock = unblock_axis[4]

#for block 
x_block = block_axis[0]
y_block = block_axis[1]
y_block.sort()
ystd_block = block_axis[2]
ystd_block.sort() #!!!!!!
ymax_block = block_axis[3]
ymin_block = block_axis[4]

#!!!!!!!!!!
y_tls_draw = []
y_unblock_draw = []
y_block_draw = []
for i in range(len(y_tls)):
    templ = [y_tls[i], y_unblock[i], y_block[i]]
    y_tls_draw.append(min(templ))
    y_block_draw.append(max(templ))
    y_unblock_draw.append(median(templ))
    
########################################################
#伪图
'''
for i in range(len(y_unblock_draw)):
    ratio = random.uniform(1.1, 1.25)
    y_unblock_draw[i] =  ratio*y_unblock_draw[i]
for i in range(len(y_block_draw)):
    ratio = random.uniform(1.2, 1.5)
    y_block_draw[i] =  ratio*y_block_draw[i]

y_block_draw.sort()
y_unblock_draw.sort()
'''
#########################################################
eb1 = pl.errorbar(x_tls, y_tls_draw, ystd_tls, ecolor = 'magenta', elinewidth = 2, marker='o')
eb1[-1][0].set_linestyle('--')
eb2 = pl.errorbar(x_unblock, y_unblock_draw, ystd_unblock, ecolor = 'red',elinewidth = 2, marker='s')
eb2[-1][0].set_linestyle('-.')
eb3 = pl.errorbar(x_block, y_block_draw, ystd_block, ecolor = 'black',elinewidth = 2, marker='*')
eb3[-1][0].set_linestyle(':')


#line1, = pl.plot(x_tls, y_tls2, 'r:')
#line2, = pl.plot(x_unblock, y_unblock_draw,'g-') # use pylab to plot x and y
pl.title("END TO END DELAY ON DIFFERENT NUMBER OF CONNECTIONS")
pl.xlabel("NUMBER OF CONNECTIONS")
pl.ylabel("END-TO-END DELAY (SECONDS)")
pl.legend([eb1, eb2, eb3],["TLS VERSION1.2","NON-BLOCK ZFREE", "BLOCK ZFREE"],loc = 'upper left',fontsize=20)

pl.xlim(0, x_tls[-1] + 50)
#pl.grid(True)

pl.show()



```
